# 💳 Luhn Algorithm 💳

The **Luhn algorithm** (also known as modulo 10 algorithm) is a checksum formula used to validate identification numbers such as credit card, IMEI, and national ID numbers. Developed by IBM scientist Hans Peter Luhn in 1954, it detects common data entry errors, ensuring numbers are structurally valid before processing.
## Algorithm
- Process digits from right to left.
- Double every second digit (starting from index 1).
- If doubling results in a number > 9, subtract 9.
- Sum all digits.
- If the total is divisible by 10 $\rightarrow$ number is valid.
***

```python
import re as regex

class LuhnValidator:
    def __init__(self, n = 16):
        self.expected_length = n
        self.vector = [(1 if (i & 1) == 0 else 2) for i in range(n)]

    def validate(self, number: str) -> bool:
        number = regex.sub(r"[\s,-]", "", number)

        if len(number) != self.expected_length:
            print("Invalid length")
            return False
        
        sumTotal = 0
        for i in range(len(number)):
            digit = int(number[len(number) - 1 - i])
            product = digit * self.vector[i]

            if product > 9:
                product -= 9

            sumTotal += product

        return sumTotal % 10 == 0
```

## Driver code

```python
validator = LuhnValidator(16)
card_numbers = ["3379 5135 6110 8795", "5103-2088-7967-2792", "5664 3210 2891 5509"]

for card in card_numbers:
    print(f"{card} is {'Valid' if validator.validate(card) else 'Invalid'}")
```
